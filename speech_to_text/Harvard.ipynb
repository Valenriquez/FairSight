```python
!pip install yt-dlp
!pip install git+https://github.com/openai/whisper.git
import os
import yt_dlp
import whisper

# Descargar el audio de YouTube
def download_audio(youtube_url, output_path="audio.mp3"):
    ydl_opts = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'outtmpl': output_path,
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([youtube_url])

# Transcribir usando Whisper
def transcribe_audio(audio_path):
    model = whisper.load_model("base")  # Carga el modelo de Whisper
    result = model.transcribe(audio_path)
    return result['text']

# Ejemplo de uso
youtube_url = 'https://youtu.be/sZ60bY2pJfo?si=LiFrUARjZtOQ2FcQ'  # URL del video
audio_path = 'audio.mp3'

# Descargar y transcribir
download_audio(youtube_url, audio_path)
transcription = transcribe_audio(audio_path)
print(transcription)
```

```python
import requests

# Endpoint de la API
url = 'http://api.openweathermap.org/data/2.5/weather'

# Parámetros de la solicitud
params = {
    'q': 'Monterrey',          # Ciudad
    'appid': 'YOUR_API_KEY',   # Tu API key obtenida en OpenWeatherMap
    'units': 'metric'          # Unidades (métricas para obtener la temperatura en Celsius)
}

# Realizar la petición GET
response = requests.get(url, params=params)

# Verificar si la respuesta fue exitosa (código 200)
if response.status_code == 200:
    # Parsear la respuesta JSON
    weather_data = response.json()

    # Extraer datos relevantes
    city = weather_data['name']
    temp = weather_data['main']['temp']
    description = weather_data['weather'][0]['description']

    print(f"Clima en {city}: {temp}°C, {description}")
else:
    print(f"Error {response.status_code}: {response.text}")
```

```python
from youtube_transcript_api import YouTubeTranscriptApi

# Función para extraer el ID del video desde el link de YouTube
def get_video_id(youtube_url):
    return youtube_url.split("v=")[-1]

# Función para obtener la transcripción
def obtener_transcripcion(youtube_url, idioma='en'):
    video_id = get_video_id(youtube_url)
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=[idioma])
        for i in transcript:
            print(f"{i['start']}: {i['text']}")
    except Exception as e:
        print(f"No se pudo obtener la transcripción: {e}")

# Ejemplo de uso
youtube_url = 'https://www.youtube.com/watch?v=rLEhL8Y4SdU'
obtener_transcripcion(youtube_url)
```

```python
from pytube import YouTube
from moviepy.editor import VideoFileClip

# Descargar el video de YouTube y extraer el audio
def descargar_audio(youtube_url, output_path='audio.mp3'):
    yt = YouTube(youtube_url)
    video = yt.streams.filter(only_audio=True).first()
    video.download(filename='video.mp4')

    # Extraer el audio usando moviepy
    video_clip = VideoFileClip('video.mp4')
    video_clip.audio.write_audiofile(output_path)

# Ejemplo de uso
youtube_url = 'https://www.youtube.com/watch?v=whZhRs9HHQI'
descargar_audio(youtube_url)
```

```python
import whisper
import pandas as pd

# Cargar el modelo de Whisper
model = whisper.load_model("base")  # Puedes usar "tiny", "small", "medium", "large" según la precisión que quieras

# Transcribir el audio
def transcribir_audio(audio_path, output_tsv):
    result = model.transcribe(audio_path, verbose=True)

    # Crear una lista para almacenar los datos de transcripción
    transcripciones = []

    # Extraer la información del resultado
    for segment in result['segments']:
        start = segment['start']
        end = segment['end']
        text = segment['text']
        transcripciones.append([start, end, text])

    # Crear un DataFrame con los datos
    df = pd.DataFrame(transcripciones, columns=['start', 'end', 'text'])

    # Guardar en un archivo TSV
    df.to_csv(output_tsv, sep='\t', index=False)


# Ejemplo de uso
audio_path = 'audio.mp3'  # Reemplaza por el nombre de tu archivo de audio
output_tsv = 'transcripcion.tsv'
transcribir_audio(audio_path, output_tsv)
```

```python
import csv
from pyannote.audio import Pipeline

# Cargar el pipeline preentrenado sin cambiar el dispositivo (por defecto usará CPU)
pipeline = Pipeline.from_pretrained(
    "pyannote/speaker-diarization-3.1",
    use_auth_token="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"
)

# Aplicar el pipeline preentrenado al archivo de audio
diarization = pipeline("audio.mp3")

# Nombre del archivo CSV de salida
csv_filename = "compressed_diarization_output.csv"

# Crear lista para almacenar los intervalos comprimidos
compressed_diarization = []

# Inicializar variables para almacenar el speaker actual y su intervalo
current_speaker = None
current_start = None
current_end = None

# Recorrer los resultados de la diarización
for turn, _, speaker in diarization.itertracks(yield_label=True):
    if speaker == current_speaker:
        # Si el mismo speaker continúa, actualiza el final del intervalo
        current_end = turn.end
    else:
        # Si el speaker cambia, guarda el intervalo anterior
        if current_speaker is not None:
            compressed_diarization.append([f"{current_speaker}", current_start, current_end])
        # Actualizar al nuevo speaker y su intervalo
        current_speaker = speaker
        current_start = turn.start
        current_end = turn.end

# Añadir el último intervalo si existe
if current_speaker is not None:
    compressed_diarization.append([f"{current_speaker}", current_start, current_end])

# Guardar los intervalos comprimidos en el archivo CSV
with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    # Escribir la cabecera
    writer.writerow(["Speaker", "Start Time", "End Time"])

    # Escribir los datos comprimidos
    writer.writerows(compressed_diarization)

print(f"Archivo CSV comprimido guardado como {csv_filename}")
```

```python
import pandas as pd

# Cargar el archivo JSON
def leer_json_a_dataframe(json_path):
    # Lee el archivo JSON
    data = pd.read_json(json_path)

    # Imprime las primeras filas del DataFrame
    print(data.head())

    return data

# Ejemplo de uso
json_path = 'time_line.json'  # Reemplaza con el nombre de tu archivo JSON
data = leer_json_a_dataframe(json_path)
```

```python
data.info()

data['speaker'] = data['speaker'].replace({
    'SPEAKER_00': 'Donald Trump',
    'SPEAKER_01': 'Kamala Harris',
    'SPEAKER_02': 'Interviewer',
})
```

```python
import matplotlib.pyplot as plt

# Crear figura
plt.figure(figsize=(10, 6))

# Dibujar las barras
for i, speaker in enumerate(data['speaker']):
    plt.barh(speaker, data['end_time'][i] - data['start_time'][i], left=data['start_time'][i])

plt.xlabel('Tiempo')
plt.ylabel('Orador')
plt.title('Líneas de tiempo por orador')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(data)
df['duration'] = df['end_time'] - df['start_time']

# Agrupar por orador y sumar las duraciones
grouped = df.groupby('speaker')['duration'].sum()

# Crear el gráfico
grouped.plot(kind='bar', stacked=True)
plt.xlabel('Orador')
plt.ylabel('Tiempo total hablado')
plt.title('Tiempo total de intervención por orador')
plt.show()
```

```python
# Crear el gráfico de pie
grouped.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.ylabel('')
plt.title('Proporción de tiempo hablado por orador')
plt.show()
```

```python
import networkx as nx

# Crear un grafo
G = nx.Graph()

# Agregar nodos (oradores)
G.add_nodes_from(df['speaker'].unique())

# Agregar aristas (interacciones entre oradores)
for i in range(1, len(df)):
    G.add_edge(df['speaker'][i-1], df['speaker'][i])

# Dibujar la red
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=3000, font_size=15, font_color='black')
plt.title('Gráfico de red de interacciones entre oradores')
plt.show()
```

```python
import numpy as np

# Crear una columna de tiempo acumulado para cada orador
df['cumulative_time'] = df.groupby('speaker')['end_time'].cumsum()

# Crear un gráfico de líneas
plt.figure(figsize=(10, 6))
for speaker, group in df.groupby('speaker'):
    plt.plot(group['start_time'], group['cumulative_time'], label=speaker)

plt.xlabel('Tiempo de inicio')
plt.ylabel('Tiempo acumulado')
plt.title('Línea de tiempo acumulada por orador')
plt.legend()
plt.show()
```
